```python
import matplotlib.pyplot as plt       
```

```python
def fun(sold,company):
    li = []
    sum1 = sum(sold)
    for i in range(len(sold)):
        num =round( (sold[i]* 100)/sum1)
        print(company[i],num)
        li.append(num)
    s = str(li)
    return s
company = ['Dell','Asus','HP','Lenovo']
sold = [1200,4000,5500,1500]
color= ['red','green','blue','darkblue']

plt.pie(sold,labels=company,colors=color,explode=[0.1,0.2,0.1,0.1],autopct=fun(sold,company),counterclock=False,startangle=90)
plt.show()
```
