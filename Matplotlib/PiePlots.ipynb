```python
import matplotlib.pyplot as plt  
```

```python
label = ['A','B','C','D']
size = [3,5,10,8]
color= ['red','green','blue','darkblue']
#adding data details in Pie Graph
plt.pie(size,colors=color, labels=label,autopct='%.3f%%')
#lets make both sides equal
plt.axis("equal")
plt.legend()
plt.title('Pie Graph')
plt.show()
```

```python
#lets say we want to highlight some particular area we can do it by using explode function
explode = [0.1,0,0,0] #it will take a array
label = ['A','B','C','D']
size = [3,5,10,8]
color= ['red','green','blue','darkblue']
#adding data details in Pie Graph
plt.pie(size,colors=color, labels=label,autopct='%.3f%%',explode=explode)
#lets make both sides equal
plt.axis("equal")
plt.title('Pie Graph')
plt.show()
```

```python
# as we can see it is currently in the anti clock wise direction so we can make it to clockwise also

explode = [0.1,0,0,0] #it will take a array
label = ['A','B','C','D']
size = [3,5,10,8]
color= ['red','green','blue','darkblue']
#adding data details in Pie Graph
plt.pie(size,colors=color, labels=label,autopct='%.3f%%',explode=explode,counterclock=False) #just type counterclock="False"
#lets make both sides equal
plt.axis("equal")
plt.title('Pie Graph')
plt.show()
```

```python
company = ['Dell','Asus','HP','Lenovo']
sold = [1200,4000,5500,1500]
color= ['red','green','blue','darkblue']

plt.pie(sold,labels=company,colors=color)
plt.legend()
plt.title('Company Sales')
plt.show()
```
