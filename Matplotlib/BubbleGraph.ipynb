```python
import matplotlib.pyplot as plt
```

```python
year = [2010,2013,2014,2015,2016,2017]
salary = [12,13,14,15,16,17]
population = [120,234,456,677,778,899]
```

```python
plt.scatter(year,salary, s=population, color='r')
```

```python
plt.scatter(year,salary, s=population,c='r', alpha=0.5)
#alpha actually increase or decrease the opacity of the bubbles
```

```python
#adding different colors to the bubbles
import numpy as np
c = np.arange(len(year))
c[1] = c[0]
```

```python
plt.scatter(year,salary,s=population, c = c, alpha=0.5)
plt.title('Bubble Graph')
plt.show()
```
