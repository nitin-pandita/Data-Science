```python
import numpy as np 
x = np.array([1,2,3,4,5])

y = x ** 3

plt.plot(x,y)
plt.show()
```

```python
#its not that curve so we will use 
import numpy as np 
# x = np.array([1,2,3,4,5])
x = np.arange(0,5,0.1)

y = x ** 3

plt.plot(x,y)
plt.show()

# now it's more curve now
```

```python
import numpy as np 
# x = np.array([1,2,3,4,5])
x = np.arange(0,5,0.1)
y = x ** 3

plt.plot(x,y,color='pink')
plt.show()
```

```python
# we can give marker and color separately
import numpy as np 
# x = np.array([1,2,3,4,5])
x = np.arange(0,5,0.1)

y = x ** 3

plt.plot(x,y,color="green", marker='p')
plt.show()
```

```python
#lets increase the line width
import numpy as np 
# x = np.array([1,2,3,4,5])
x = np.arange(0,5,0.1)

y = x ** 3

plt.plot(x,y, color="green",linewidth= 5)
plt.show()
```

```python
import numpy as np 
# x = np.array([1,2,3,4,5])
x = np.arange(0,5,0.1)

y = x ** 3
y2 = x ** 2

plt.plot(x,y, color="green", label='x**3') #label are used when calling a label function
plt.plot(x,y2, color="blue", label='x**2')
plt.xlabel('Sales')
plt.ylabel('Loss in Business')
# we show the dataset info we can use legend
plt.legend()
# we can also create a grid for the graph
plt.grid()
# what if we want to add some text in the graph
plt.text(3.3,80,"Green line", fontsize=15)
plt.title('Graph Plotting')
plt.show()
```
