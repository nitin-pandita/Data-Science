```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x,y,z)
```

```python
#lets try to make a cos function
plt.plot(x,y)
plt.title('Sin Wave')
plt.ylabel('Sin')
plt.xlabel('Numbers')
plt.show()
```

```python
#let plot the cosin function
plt.plot(x,z)
plt.title('Cosin Wave')
plt.ylabel('Cos')
plt.xlabel('Sin')
plt.show()
```

```python
#lets make a hyperbola
x = np.linspace(-10,10,100)
y = x ** 2
plt.title('Hyperbola')
plt.xlabel('x**2')
plt.ylabel('x')
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-5,5,100)
plt.plot(x,np.sin(x), 'r.')
plt.plot(x,np.cos(x), 'b.')
plt.show()
```

### Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
subjects = ['Hindi','English','Math','Science','Geography']
students = [45,56,23,66,23]

ax.bar(subjects,students)
plt.xlabel('Subjects')
plt.ylabel('Students')
plt.title('Favourite subject')
plt.show()
```

### Pie Chart

```python
languages = ['Hindi','English','German','Spanish','French']
interested = [34,66,22,64,78]

plt.pie(interested,labels=languages, autopct='%1.1f%%')
plt.show()
```

### 3D Scater Plot


```python
from matplotlib import colors
fig = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)

ax.scatter(x,y,z)
plt.show()
```
